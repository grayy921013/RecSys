```python
# STEP: 1.2 BM25F

# import Projects.scikit_learn.sklearn as a
from time import time
from Bm25 import Bm25Vectorizer
# from sklearn.feature_extraction.text import Bm25Vectorizer

# from vectorizers import TfidfVectorizer as BM25Vectorizer

def index_data(data):
    t0 = time()
    vectorizer = Bm25Vectorizer(max_df=0.5, stop_words='english')
    X_train = vectorizer.fit_transform(data)
    duration = time() - t0
    print("n_samples: %d, n_features: %d" % X_train.shape)
    print("duration: %d\n" % duration)
    return X_train, vectorizer

indexes = []
vectorizers = []
i = 0
for d in data:
    i += 1
    print ' Processing Dataset %d ' % i
    
    index, vectorizer = index_data(d)
    indexes.append(index)
    vectorizers.append(vectorizer)
```

```python
# STEP: 2.2 Scoring - BM25F
scores = []
for index in indexes:
    t0 = time()
    score = index.dot(index.T)
    duration = time() - t0
    # np.fill_diagonal(score, 0)
    print("n_movies: %d, n_related_movies: %d" % score.shape)
    print("duration: %f\n" % duration)
    scores.append(score)
```

```python
# STEP: 2.3 Scoring - JACCARD
from scipy.sparse import csr_matrix

def vectorized_union(intersection, data):
    
    # Step 1: Aggregate each row
    row = np.ravel(data.sum(1))
    
    # Step 2: Append a row/column full of ones
    ones = np.ones(row.shape)
    matrix_1 = np.append([row], [ones], axis=0)
    matrix_2 = np.append([ones.T], [row.T], axis=0).T
    
    # Step 3: Calculate the sum of each possible combination
    union = matrix_2.dot(matrix_1)
    print 'union.shape', union.shape
    # np.fill_diagonal(union, 0)  # Remove diagonal?
    
    # Step 4: Remove intersection
    # Step 4.1: 0 everything that does not have an intersection
    #           This step is not for UNION, but in JACCARD we dont care about 
    #           Cells without intersection
    sparse_union = csr_matrix((union[intersection.nonzero()], intersection.nonzero()), shape=intersection.shape)
    
    # Step 4.2: Remove Intersection
    sparse_union.data = sparse_union.data - intersection.data
    
    return sparse_union

scoresJ = []
for index in indexes:
    t0 = time()
    binary_index = index
    binary_index[binary_index != 0] = 1
    
    intersection = binary_index.dot(binary_index.T)
    union = vectorized_union(intersection, binary_index)
    result = union.copy()
    result.data = np.divide(intersection.data, union.data)

    duration = time() - t0
    print("n_movies: %d, n_related_movies: %d" % intersection.shape)
    print("duration: %f\n" % duration)
    scoresJ.append(intersection)
```

```python
# STEP: 3.0 Aggregate
# # Analyze how to solve the memory problem

# c_title_field
# c_year_field
# c_genres_field
# c_tags_field
weights = [0.45, 0.05, 0.05, 0.45]

aggregated_data = weights[0]*scores[0]
for i in range(1,len(scores)):
    aggregated_data += weights[i]*scores[i]

# aggregated_data = weights[0]*scoresJ[0]
# for i in range(1,len(scoresJ)):
#     aggregated_data += weights[i]*scoresJ[i]
    
# aggregated_data = scores[3]
print "Final model %d " % aggregated_data.shape[0]
```

```python
# STEP: 4 TOP
import numpy as np
TOP_N = 21

MOVIES = aggregated_data.shape[0]

top = []
j = 0
for i in xrange(MOVIES):
    if i % 10000 == 0:
        print i
    
    # Only get the data
    related_movies_scores_for_i = aggregated_data[i,:].data
    related_movies_id_for_i = aggregated_data[i,:].indices
    
    # Look the top N values in that N
    if len(related_movies_id_for_i) < TOP_N:
        # If it already only has fewer possible similars, just pick the whole set
        top_n_ids = related_movies_id_for_i
        j += 1
    else:
        # Split the whole thing
        top_n = np.argpartition(related_movies_scores_for_i, -TOP_N, axis=0)[-TOP_N:]
        top_n_ids = related_movies_id_for_i[top_n]
    
    r = set()
    for i in top_n_ids:
        r.add(c_id_field[i])
    top.append(r)
print 'Movies Processed: %d Movies without enough Related Movies: %d' % (len(top), j)

# Reference:
# https://stackoverflow.com/questions/6910641/how-to-get-indices-of-n-maximum-values-in-a-numpy-array
```
